```python
# %load C:\Users\Abhisekh\Desktop\Python_basics\Exercises\Assignments\Third_question.py
#Write generator to create prime number series

def is_prime(n):
    if (n <= 1):
        return False
    if (n == 2):
        return True
    if (n % 2 == 0):
        return False

    i = 3
    while i <= n/2:
        if n % i == 0:
            return False
        i = i + 2

    return True


def prime_generator():
    n = 1
    while True:
        n += 1
        if is_prime(n):
            yield n


generator = prime_generator()

for i in range(5):
    print(next(generator))
```

```python
# %load C:\Users\Abhisekh\Desktop\Python_basics\Exercises\Assignments\Sixth_question.py
#Program to accept the strings which contains all vowels

def check_vowel(string):
    str = string.lower()

    vowels = set("aeiou")

    val = set()

    for char in str:
        if char in vowels:
            val.add(char)
        else:
            pass

    if len(val) == len(vowels):
        print("string is accepted")
    else:
        print("string is not accepted")

string =input("enter a string")
check_vowel(string)
```

```python
# %load C:\Users\Abhisekh\Desktop\Python_basics\Exercises\Assignments\Seventh_question.py
#Check if input number is Palindrome numbers or not



def check_Palindrome_num(num):
    temp = num
    rev = 0
    while num>0:
        dig =num%10
        rev =rev*10+dig
        num =num//10

    if temp==rev:
        print("palindrome")
    else:
        print("not palidrome")

num = int(input("enter a number "))
check_Palindrome_num(num)
```

```python
# %load C:\Users\Abhisekh\Desktop\Python_basics\Exercises\Assignments\Eighth_question.py
#Create Custom decorator to get function execution time (add differnt sleep values in differnt functions)


import time

def func_execution_time(func):

    def inner(*args, **kwargs):
        start = time.time()

        func(*args, **kwargs)

        end = time.time()

        print("time taken by ", func.__name__, end-start)
    return inner

@func_execution_time
def sum(a,b):
    c=a+b
    print(c)
    time.sleep(3)

sum(34,56)

@func_execution_time
def div(x,y):
    z =x/y
    print(z)
    time.sleep(2)

div(20,4)
```

```python
# %load C:\Users\Abhisekh\Desktop\Python_basics\Exercises\Assignments\Ninth_question.py
# Use Generator to parse csv file

import csv

path = 'C:\\Users\\Abhisekh\\Desktop\\Python_basics\\Exercises\\Assignments\\text.csv'

def parse_Csv(file):
    with open(file, "r") as csv1:
        reader1 = csv.reader(csv1)

        for row in reader1:
            yield row


for i in parse_Csv(path):
    print(i)
```
